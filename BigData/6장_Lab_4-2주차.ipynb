## Lab 7 : 파생변수 추가하기
- df.assign(컬럼이름 = 함수)

```python
import pandas as pd
exam = pd.read_csv('exam.csv')
exam.head(3)
```

```python
# math, english, science 값을 합산한 값을 갖는 파생변수 total 생성
exam.assign(total = exam['math'] + exam['english'] + exam['science'])
exam.head()
```

```python
# math, english, science 값을 합산한 값을 갖는 파생변수 total 생성
# math, english, science 값을 평균한 값을 갖는 파생변수 mean 생성

exam.assign(
    total = exam['math'] + exam['english'] + exam['science'],
    mean = (exam['math'] + exam['english'] + exam['science']) / 3

)
exam.head()
```

## Lav 7 (혼자서 해보기) : mpg 데이터를 이용하여 문제 분석
- Q1 : mpg 사본을 만들고, cty와 hwy를 더한 '합산 연비 변수' totaldmf 추가 하시오

```python
mpg = pd.read_csv('mpg.csv')
```

```python
mpg_new = mpg.copy()
```

```python
mpg_new = mpg_new.assign(total = mpg_new['cty'] + mpg_new['hwy'])
mpg_new.head(3)
```

- Q2 : 합산 연비를 total을 2로 나누어 평균 연비 변수(mean)를 추가하시오

```python
mpg_new = mpg_new.assign(mean = mpg_new['total'] / 2)
mpg_new.head(3)
```

- Q3 : 평균 연비가 가장 높은 자동차 3종의 데이터를 출력하시오

```python
mpg_new.sort_values('mean', ascending=False).head(3)
```
