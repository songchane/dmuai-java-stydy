### 직업별 월급 평균 및 남여 종사자 수 표 만들기

```python
import seaborn as sns
```

```python
import pandas as pd

welfare = pd.read_spss('Koweps_hpwc14_2019_beta2.sav')
w = welfare.copy()
```

```python
w = w.rename(
    columns={'h14_g3' : 'sex', 
             'h14_g4' : 'birth',
             'h14_g10': 'marriage_type',
             'h14_g11':'religion',
             'p1402_8aq1':'income',
             'h14_eco9':'code_job',
             'h14_reg7':'code_region'}
)
```

```python
list_job = pd.read_excel('Koweps_Codebook_2019.xlsx',sheet_name='직종코드')
```

```python
w =w.merge(list_job,how='left',on='code_job')
```

```python
w.dropna(subset=['code_job'])[['code_job','job']].head()
```

```python
job_income = w.dropna(subset = ['job','income'])\
            .groupby('job',as_index=False)\
            .agg(mean_income = ('income','mean'))
job_income
```

```python
import numpy as np
w['sex'] = np.where(w['sex']==1,'male','female')
```

```python
job_male = w.dropna(subset = ['job'])\
            .query('sex=="male"')\
            .groupby('job',as_index = False)\
            .agg(male_n=('job','count'))
job_male
```

```python
job_female = w.dropna(subset = ['job'])\
            .query('sex=="female"')\
            .groupby('job',as_index = False)\
            .agg(female_n=('job','count'))
job_female
```

```python
df = job_income.merge(job_male, how='left', on='job')
df = df.merge(job_female, how='left', on='job')
df
```
